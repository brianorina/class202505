# 5. Visualize the Data

```python
# Visualize the distribution of the median house value using a histogram with a KDE (smoothed line).
# This helps us identify skewness, concentration, and outliers in the target variable.

plt.figure(figsize=(8, 5))
sns.histplot(housing_df['median_house_value'], bins=50, kde=True)
plt.xlabel('Median House Value')
plt.ylabel('Number of Districts')
plt.title('Distribution of Median House Value with KDE')
plt.show()
```

```python
# Visualize the frequency distribution of the 'ocean_proximity' categorical feature.
# This helps us see which categories are most represented and detect any class imbalance.

plt.figure(figsize=(8, 5))
sns.countplot(x='ocean_proximity', data=housing_df)
plt.title('Frequency Distribution of Ocean Proximity')
plt.xlabel('Ocean Proximity')
plt.ylabel('Number of Districts')
plt.xticks(rotation=60)
plt.show()
```

```python
# Display a correlation matrix heatmap for all numerical features.
# This visual summary helps identify which features are strongly correlated, either positively or negatively.
# Values closer to 1 or -1 indicate stronger linear relationships.

plt.figure(figsize=(10, 8))
corr_matrix = housing_df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap for Numerical Features')
plt.show()
```

## Visualization for Train Data

```python
# Survival Distribution
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Survived', data=titanic_df)
plt.title('Survival Distribution')
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')

# Add count labels on top of the bars
ax.bar_label(ax.containers[0])

plt.show()
```

```python
# Survival by Sex
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Survived', hue='Sex', data=titanic_df)
plt.title('Survival by Sex')
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
total = len(titanic_df)
# Use ax.bar_label() with a custom formatter for percentages
def percent_formatter(x):
    return f'{100 * x / total:.1f}%'

# Add the percentage labels to each bar
ax.bar_label(ax.containers[0], labels=[percent_formatter(rect.get_height()) for rect in ax.containers[0]])
ax.bar_label(ax.containers[1], labels=[percent_formatter(rect.get_height()) for rect in ax.containers[1]])

plt.show()
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(8, 5))
sns.histplot(titanic_df['Age'], bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# Correlation Heatmap (numeric only)

# Step 1: Create a figure with an appropriate size (e.g., width=10, height=6).
plt.figure(figsize=(10, 6))

# Step 2: Generate a correlation matrix using the .corr() method.
# Make sure to include only numeric columns using numeric_only=True.
corr_matrix = titanic_df.corr(numeric_only=True)

# Step 3: Use seaborn's heatmap to visualize the correlation matrix.
# Enable annotations (annot=True) and apply a diverging color map like 'coolwarm'.
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Step 4: Add a descriptive title to the plot.
plt.title('Correlation Heatmap of Titanic Numeric Features')

# Step 5: Display the plot.
plt.show()
```
